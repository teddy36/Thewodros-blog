```python
import seaborn as sns
```

```python
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
sns.histplot(tips['total_bill']) # distplot -- distribution plot
```

```python
sns.kdeplot(tips['total_bill'])   # # kde--means--- A kernel density estimate (KDE) plot
```

```python
sns.distplot(tips['total_bill'],kde=False) # kde--means--- A kernel density estimate (KDE) plot
```

```python
sns.distplot(tips['total_bill'],kde=False,bins=40)
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips) #kind=scatter plot--(joint plot passing two columns(x,y,and data)arg.)
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='hex')  # hex --- kind of hexagon plot
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='reg')  # reg--means regration plot
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='kde') # kde-- is kind of two dimention plot
```

```python
sns.pairplot(tips)
```

```python
sns.pairplot(tips,hue='sex') # hue--help us to see just a category
```

```python
sns.pairplot(tips,hue='sex',palette='coolwarm')  # palette--will help us to choose the kind of color for a parameter
```

```python
sns.rugplot(tips['total_bill']) # reg -- a dash mark plot
```
